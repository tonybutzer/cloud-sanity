```python
# https://snorfalorpagus.net/blog/2016/07/15/masking-netcdf-datasets-using-rasterio/
```

```python
with rasterio.open(redfile) as src:
    print(src.profile)
    # show(src, 1, cmap='Reds')
    print(type(src))
    print(dir(src))
    print(src.shape)
    # rasterio.plot.show(src)
    print(src.xy(0,0))
    my_raster = src.read(1)
```
